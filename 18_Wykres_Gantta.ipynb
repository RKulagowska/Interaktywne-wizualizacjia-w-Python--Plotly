```python
import plotly.figure_factory as ff
df = [dict(Task="Project A", Start="2019-01-01", Finish = "2020-12-12"),
      dict(Task="Project B", Start="2019-03-01", Finish = "2019-10-12"),
      dict(Task="Project C", Start="2019-02-01", Finish = "2020-07-12"),
      dict(Task="Project D", Start="2019-04-01", Finish = "2020-03-12")]       
ff.create_gantt(df)
```
