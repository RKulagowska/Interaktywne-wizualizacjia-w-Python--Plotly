```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

np.linspace rozkłada na 100 równych danych z przedziału od 0 do 20 
np.random.rand(100,4) daje 100 próbek i 4 zmiennne ( wymiar 100 x 4) 

```python
x = np.linspace(0,20,100)
y = np.random.rand(100,4)
y
```

np.cumsum(y, axis= 0 ) suma skumulowana z y po kolumnach

```python
y = np.cumsum(y,axis=0)
y
```

```python
plt.plot(x,y)
```

```python
sns.set()
plt.plot(x,y)
```
