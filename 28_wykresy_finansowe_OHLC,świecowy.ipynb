```python
import plotly.express as px

def fetch_finacial_data(company='AMZN'):
  import pandas_datareader.data as web
  return web.DataReader(name=company,data_source='stooq')
df = fetch_finacial_data()
df = df.reset_index()
df.head(2)
```

```python
df = df[df.Date > '2022-01-01']
df.describe()
```

```python
import plotly.graph_objects as go
```

```python
fig= go.Figure(
    go.Ohlc(
        x= df.Date,
        open = df.Open,
        high = df.Close,
        low = df.Low,
        close = df.Close

    )
)
fig.show()
```

```python
fig= go.Figure(
    go.Candlestick(
        x= df.Date,
        open = df.Open,
        high = df.Close,
        low = df.Low,
        close = df.Close

    )
)
fig.show()
```
