```python
def fetch_finacial_data(company='AMZN'):
  import pandas_datareader.data as web
  return web.DataReader(name=company,data_source='stooq')

df_amnz = fetch_finacial_data(company= 'AMZN')
df_aapl = fetch_finacial_data(company= 'AAPL')
df_msft = fetch_finacial_data(company= 'MSFT')
df_nflx= fetch_finacial_data(company= 'NFLX')
df_twtr = fetch_finacial_data(company= 'TWTR')

df_amnz.head(2)
```

łączenie obiektóz za pomocą pandas 

```python
import pandas as pd

df = pd.concat([df_amnz.Close,df_aapl.Close, df_msft.Close, df_nflx.Close, df_twtr.Close], axis=1, join='inner')
df.head(2)
```

```python
import pandas as pd

df = pd.concat([df_amnz.Close,df_aapl.Close, df_msft.Close, df_nflx.Close, df_twtr.Close], axis=1, join='inner')
df.columns = ['amazon', 'apple' , 'microsoft', 'netflix', 'twitter']
df= df.reset_index()
df.head(2)
```

```python
df.describe()
```

```python
import plotly.graph_objects as go

trace0 = go.Scatter(x=df.Date,
                    y= df.amazon,
                    name = 'Amazon')

data=[trace0]

fig = go.Figure(data= data)
fig.show()
```

```python
import plotly.graph_objects as go

trace0 = go.Scatter(x=df.Date,
                    y= df.amazon,
                    name = 'Amazon')

data=[trace0]
layout= go.Layout(yaxis_type = 'log')

fig = go.Figure(data= data, layout = layout)
fig.show()
```

```python
import plotly.graph_objects as go

trace0 = go.Scatter(x=df.Date,
                    y= df.amazon,
                    name = 'Amazon')
trace1 = go.Scatter(x=df.Date,
                    y= df.apple,
                    name = 'Apple')
trace2 = go.Scatter(x=df.Date,
                    y= df.microsoft,
                    name = 'Microsoft')
trace3 = go.Scatter(x=df.Date,
                    y= df.netflix,
                    name = 'Netflix')
trace4 = go.Scatter(x=df.Date,
                    y= df.twitter,
                    name = 'Twitter')
data= [trace0, trace1, trace2 ,trace3, trace4]
layout= go.Layout(yaxis_type = 'log')

fig = go.Figure(data= data, layout = layout)
fig.show()
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(
    x=df.Date,
    y= df.amazon,
    name='Amazon'
))
fig.show()
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(
    x=df.Date,
    y= df.amazon,
    name='Amazon'
))
fig.add_trace(go.Scatter(
    x=df.Date,
    y= df.apple,
    name='Apple'
))
fig.show()
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(
    x=df.Date,
    y= df.amazon,
    name='Amazon'
))
fig.add_trace(go.Scatter(
    x=df.Date,
    y= df.apple,
    name='Apple'
))
fig.update_layout(yaxis_type = 'log')
fig.show()
```
