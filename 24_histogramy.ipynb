```python
from seaborn import load_dataset
import plotly.express as px
```

```python
df= load_dataset('mpg')
df.head()
```

```python
df.describe()
```

```python
df.origin.value_counts()
```

```python
px.histogram(df,x='mpg')
```

```python
px.histogram(df,x='mpg',nbins= 50, title="Rozkład zmiennej mpg")
```

wartość prawdopodobieństwa

```python
px.histogram(df,x='mpg',nbins= 50, title="Rozkład zmiennej mpg", histnorm ='probability density')
```

```python
px.histogram(df,x='mpg',nbins= 50, title="Rozkład zmiennej mpg", histnorm ='probability density',
             facet_row='origin')
```

```python
px.histogram(df,x='mpg',nbins= 50, title="Rozkład zmiennej mpg", histnorm ='probability density',
             facet_row='origin', color='origin')
```

```python
dff= df[(df.origin == 'usa')|(df.origin =='europe')]
px.histogram(dff,x='mpg',nbins= 50, color='origin')
```

```python
px.histogram(dff,x='mpg',nbins= 50, color='origin', marginal= 'rug', title = 'Rozkład zmiennej mpg(usa vs europe', opacity=0.7)
```

```python
px.histogram(dff,x='mpg',nbins= 50, color='origin', marginal= 'box', title = 'Rozkład zmiennej mpg(usa vs europe', opacity=0.7)
```

```python
px.histogram(dff,x='mpg',nbins= 50, color='origin', marginal= 'violin', title = 'Rozkład zmiennej mpg(usa vs europe', opacity=0.7)
```

```python
px.histogram(df,x='origin')
```

distribution plot

```python
import plotly.figure_factory as ff
fig = ff.create_distplot([df.mpg], group_labels=['mpg'])
fig.show()
```

```python
fig = ff.create_distplot([df.mpg, df.acceleration], group_labels=['mpg','Acceleration'])
fig.show()
```

```python
import numpy as np

x1= np.random.randn(300)
x2= np.random.randn(300)+2

ff.create_distplot([x1,x2],['Rozkład Normalny ~ N(0,1)','Rozkład Normalny `N(2,1)'], curve_type = 'normal')
```

```python
ff.create_distplot([x1,x2],['Rozkład Normalny ~ N(0,1)','Rozkład Normalny `N(2,1)'], curve_type = 'normal', bin_size=0.2)
```

```python
x1= np.random.randn(1000)
x2= np.random.randn(1000)+2
ff.create_distplot([x1,x2],['Rozkład Normalny ~ N(0,1)','Rozkład Normalny `N(2,1)'], curve_type = 'normal', bin_size=0.2,
                   colors= ['slategray', 'magenta'])
```
