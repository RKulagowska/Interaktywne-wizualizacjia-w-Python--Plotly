```python
for template in ['plotly', 'plotly_white','plotly_dark', 'ggplot2','seaborn']:
  x= px.scatter(df.query("year ==2007"),
           x= 'gdpPercap',
           y= 'lifeExp',
           size= 'pop',
           color= 'continent',
           log_x=True, 
           size_max=50, 
           height= 400, 
           title= f'Template {template}',
           template = template)
  x.show()
```

```python
import plotly.io as pio
pio.templates.default = 'plotly_white'
df = px.data.iris()
fig = px.scatter(df, x='sepal_length', y= 'sepal_width')
fig.show()
```
