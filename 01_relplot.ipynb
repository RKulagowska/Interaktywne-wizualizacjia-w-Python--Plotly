```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df= sns.load_dataset('fmri')
df.head(3)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.describe(include=['object'])
```

sns.relplot - pokazuje dokumentacje 

```python
sns.relplot(data = df, x = 'timepoint', y= 'signal')
```

```python
sns.relplot(data = df, x = 'timepoint', y= 'signal', kind='line')
```

ci - przedzia ufnoci 

```python
sns.relplot(data = df, x = 'timepoint', y= 'signal', kind='line', ci=None)
```

```python
df.timepoint.value_counts()
```

urednienie sygnau po wartoci timepoint

```python
df.groupby('timepoint').agg('mean')
```

```python
sns.relplot(data=df, x='timepoint', y='signal')
```

```python
sns.relplot(data=df, x='timepoint', y='signal', hue= 'event')
```

```python
sns.relplot(data=df, x='timepoint', y='signal', hue= 'event', kind= 'line')
```

```python
sns.relplot(data=df, x='timepoint', y='signal', hue= 'event', kind= 'line', style= 'event')
```

```python
sns.relplot(data=df, x='timepoint', y='signal', hue= 'event', kind= 'line', style= 'event', markers=True)
```

```python
sns.relplot(data=df, x='timepoint', y='signal', hue= 'event', kind= 'line', style= 'event', markers=True, col='region')
```

```python
df.subject.value_counts()
```

```python
sns.relplot(data=df, x='timepoint', y='signal', hue= 'event', kind= 'line', style= 'event', markers=True, col='subject',col_wrap=4)
```
